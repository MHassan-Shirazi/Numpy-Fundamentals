```python
import numpy as np 
```

Numpy indexing and slicing of vectors, boolean indexing

```python
arr = np.arange(11,21)
arr
```

```python
arr[6] # indexing
```

```python
arr[1:5] # slicing
```

numpy indexing and slicing of matrix

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[3]
```

```python
arr[3,2] # indexing
```

```python
arr[0:2,1:3] # slicing
```

```python
arr[3:,3:] #slicing
```
